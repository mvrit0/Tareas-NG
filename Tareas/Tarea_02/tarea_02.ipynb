## Problema 1
Se utiliza **for** para secuenciar todos los números dentro del rango -100, 201. Con **print** se imprime cada valor que toma la variable *num*, o sea, todos los números entre -100 y 200.

```python
for num in range(-100, 201):
    print(num)
```

## Problema 2
En este problema, usamos el mismo código para secuenciar el mismo rango. Sin embargo, ahora debemos solo imprimir los números del rango divisibles por 6. Para esto, hacemos uso del condicional **if** seguido de la hint dada **%** para solo imprimir los números de la secuencia que, divididos por 6, dan resto 0.


```python
for num in range(-100, 201):
    if num % 6 == 0:
        print(num)
```

## Problema 3
Para este problema, se le pide al usuario que ingrese dos valores con la función **input** y se le acuñan dos variables *primer_numero* y *segundo_numero*. Luego se suman mediante **+** y se le acuña la variable suma. Luego se crea la variable *mensaje* con una cadena vacía para que tome valor con los condicionales posteriores. Finalmente se asigna el primer condicional mediante **if** y corresponde a si el valor de suma es menor que 200, *mensaje* muestra *es menor a 200*. Para agregar un segundo condicional se usa **elif**, y este corresponde a si el valor de suma es mayor que 200 y menor que 250, *mensaje* muestra *es mayor a 200 y menor a 250*. Finalmente, se usa **else**, para que si suma no cumple ninguno de los condicionales anteriores, *mensaje* muestre *es mayor a 250*.



```python
primer_numero = int(input("Ingresa el primer número: "))
segundo_numero = int(input("Ingresa el segundo número: "))

suma = primer_numero + segundo_numero

mensaje = ""
if suma < 200:
    mensaje = "es menor a 200"
elif suma > 200 and suma < 250:
    mensaje = "es mayor a 200 y menor a 250"
else:
    mensaje = "es mayor a 250"

print(mensaje)
```

## Problema 4
Se asignaron las variables edad y gusto_musica_urbana con sus respectivas preguntas. Luego mediante **if** y **else**, se determina si edad es menor o igual/mayor que 18, y de acuerdo a esos valores, se muestra un mensaje bajo la variable *resultado*.

```python
edad = int(input("¿Cuántos años tienes?"))
gusto_musica_urbana = input("¿Te gusta la música urbana? 'sí' o 'no': ").lower()

if edad >= 18:
    if gusto_musica_urbana == 'sí':
        resultado = "Eres mayor de edad y te gusta la música urbana"
    else:
        resultado = "Eres mayor de edad pero no te gusta la música urbana"
else:
    if gusto_musica_urbana == 'sí':
        resultado = "Eres menor de edad y te gusta la música urbana"
    else:
        resultado = "Eres menor de edad y no te gusta la música urbana"

print(resultado)



```
